```python
  # trabalho 6 - versão otimizada
import matplotlib.pyplot as plt
import numpy as np
import cv2
from google.colab import drive
from google.colab.patches import cv2_imshow
video = cv2.VideoCapture('/content/drive/MyDrive/Trabalho6/video_carro.mp4')
drive.mount('/content/drive')
```

```python
def paintLanes(img):
    color = ('b', 'g', 'r')
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            b, g, r = img[i, j]

            if (b > 180 and g > 160 and r > 160): # branco
                img[i, j] = (0, 255, 0)
            elif (b > 60 and b < 130 and g > 130 and g < 200 and r > 150 and r < 220): # amarelo
                img[i, j] = (90, 90, 0)
            else:
                img[i, j] = (1, 1, 1)

    return img

def normalWay(video):
  while(video.isOpened()):

    _, frame = video.read()

    cv2_imshow(paintLanes(frame)) # imagem inteira

def optimizedWay(video):
  while(video.isOpened()):

    _, frame = video.read()

    cv2_imshow(paintLanes(frame[frame.shape[0]-230:frame.shape[0], 0: frame.shape[1]-50])) # imagem reduzida
```

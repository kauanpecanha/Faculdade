```python
def showImages(listImages): # função de exibição de todas as imagens
  for i in range (0, len(listImages)):
    cv2_imshow(listImages[i])

def getImageInfo(listImages): # função de extração de informações das imagens, bem como criação de um array cópia para uso
  listInfo = []
  copies = listImages.copy()

  for i in range(0, len(listImages)):
    listInfo.append([listImages[i].shape[0], listImages[i].shape[1]])

  return listInfo, copies

infos, cps = getImageInfo([img1, img2, img3, img4, img5, img6])
# print(infos)s

def getHist(listImages): # função para definir os histogramas de cada uma das seis imagens
    color = ('b', 'g', 'r')
    for i in range(len(listImages)):
        plt.figure(figsize=(10, 6))

        for j, col in enumerate(color):
            hist = cv2.calcHist([listImages[i]], [j], None, [256], [0, 256])
            plt.plot(hist, color=col)
            plt.xlim([0, 256])

        plt.title(f'Histogra da imagem')
        plt.xlabel('Valor')
        plt.ylabel('Frequência')
        plt.legend(['Azul', 'Verde', 'Vermelho'])
        plt.show()

def showInfo(image):
  color = ('b', 'g', 'r')

  cv2_imshow(image)
  plt.figure(figsize=(10, 6))

  for j, col in enumerate(color):
    hist = cv2.calcHist([image], [j], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])

  plt.title(f'Histogra da imagem')
  plt.xlabel('Valor')
  plt.ylabel('Frequência')
  plt.legend(['Azul', 'Verde', 'Vermelho'])
  plt.show()
  print('\n')

def changeImages(listImages):
  for i in range(0, len(listImages)):
    for j in range(floor(listImages[i].shape[0]/2), listImages[i].shape[0]):
      for k in range(listImages[i].shape[1]):
        b, g, r = listImages[i][j, k]

        if b > 150 and g > 170 and r > 170: listImages[i][j, k] = (45, 130, 30) # listas brancas - aprovadas
        if(b < 110 and g > 110 and r > 110): listImages[i][j, k] = (100, 45, 15) # listas amarelas
        else: pass

  return listImages
```
